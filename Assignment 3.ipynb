<h1>Bit Plane Slicing</h1>

<h4> Bits 8 and 7 </h4>

```python
#Load Image
img = Image.open("DIP3E_Original_Images_CH01/Fig0115(b)(100-dollars).tif")
```

```python
#Preview
img
```

```python
#convert image to 2D array
img_ar = np.asarray(img)
```

```python
#taking bit slices of image
mask8 = img_ar & (1 << 7)
mask7 = img_ar & (1 << 6)

#creating image from sum of bit slices
bk7 = Image.fromarray(mask7 + mask8)
```

```python
bk7
```

<h4> Bits 8,7 and 6 </h4>

```python
mask6 = img_ar & (1 << 5)
bk6 = Image.fromarray(mask8 + mask7 + mask6)
```

```python
bk6
```

<h4> Bits 8,7,6 and 5 </h4>

```python
mask5 = img_ar & (1 << 4)
bk5 = Image.fromarray(mask8 + mask7 + mask6 + mask5)
```

```python
bk5
```

<h4> Bits 1,2,3 and 4 </h4>

```python
mask1 = img_ar & 1
mask2 = img_ar & (1 << 1)
mask3 = img_ar & (1 << 2)
mask4 = img_ar & (1 << 3)
bk4 = Image.fromarray(mask1 + mask2 + mask3 + mask4)
```

```python
bk4
```

<h3> Analysis </h3>

The higher order bit planes contain the most relevant information of any image. The sum of the higher order bits is enough to reconstruct the image. 

Adding lower order bits adds little information to the image and it generally contains more noise. 

This fact is evident from the excercise performed above as we were able to obtain a clear image with only bit planes 8 and 7. Adding bit plane 6 did add some information but later additions did'nt do much to improve image quality. Finally, by combining the lower order bits, we could not obtain anything more thank a black image. This proves that lower order bits don't contain much information about the image.

This technique can be used in image compression and help to reduce image size with negligible degradation in image quality.
