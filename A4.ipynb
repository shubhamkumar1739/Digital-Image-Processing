```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mul(image, ftr) :
    p_img = np.pad(image, ftr.shape[0] // 2) 
    fin_image = np.zeros((image.shape[0], image.shape[1]), dtype = np.uint8)
    for i in range(image.shape[0]) :
        for j in range(image.shape[1]) :
            fin_image[i][j] = np.sum(p_img[i : i + ftr.shape[0], j : j + ftr.shape[0]] * ftr)
    return fin_image
```

```python
img = plt.imread("DIP3E_Original_Images_CH02/Fig0219(rose1024).tif")
ar = np.asarray(img)
```

```python
plt.imshow(img)
```

```python
#Robert's Filter 
ftr = np.asarray([[1, 0],
                  [0, -1]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
ftr = np.asarray([[0, -1],
                      [1, 0]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
#Sobel's filter
ftr = np.asarray([[-1, -2, -1],
                  [0, 0, 0],
                  [1, 2, 1]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
ftr = np.asarray([[-1, 0, 1],
                  [-2, 0, 2],
                  [-1, 0, 1]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
#Laplacian filter
ftr = np.asarray([[0, 1, 0],
                  [1, -4, 1],
                  [0, 1, 0]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
ftr = np.asarray([[1, 1, 1],
                  [1, -8, 1],
                  [1, 1, 1]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
ftr = np.asarray([[0, -1, 0],
                  [-1, 4, -1],
                  [0, -1, 0]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```

```python
ftr = np.asarray([[-1, -1, -1],
                  [-1, 8, -1],
                  [-1, -1, -1]])
p_img = mul(ar, ftr)
plt.imshow(p_img)
```
